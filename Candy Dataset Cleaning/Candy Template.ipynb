# cleaning

### 4- drop all the useless columns and  rows with too much NaN , 
> tip : useless columns can be like `Internal ID`and can be also columns with too much NaN 
- rename columns for easy access

```python
df.drop(['Internal ID' ,'Q7: JOY OTHER', 'Q8: DESPAIR OTHER' ,'Unnamed: 113' , 'Q12: MEDIA [Daily Dish]' 
         , 'Q12: MEDIA [ESPN]' , 'Q12: MEDIA [Yahoo]' , 'Q12: MEDIA [Science]' 
         ,'Click Coordinates (x, y)' , 'Q9: OTHER COMMENTS'  ] ,axis = 1, inplace=True)
```

```python
for col in df.columns:
    new_col_name = col.replace('Q6 | ','')
    df.rename(columns={col : new_col_name}, inplace=True)
df.rename(columns={'Q1: GOING OUT?':'GOING OUT','Q2: GENDER':'GENDER','Q3: AGE':'AGE','Q4: COUNTRY':'COUNTRY',
                   'Q5: STATE, PROVINCE, COUNTY, ETC':'AREA','Q10: DRESS':'DRESS',
                   'Q11: DAY':'DAY'},inplace=True)
df.columns = df.columns.str.lower().str.replace(" " , "_")
df.columns
```

### 5 - clean `going out` column 
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
# print(df['going_out'].unique())
# print(df['going_out'].value_counts())
# print(df['going_out'].isna().sum())

df['going_out'] = df['going_out'].fillna(df['going_out'].mode()[0]).astype("category")

# print(df['going_out'].unique())
# df['going_out'].value_counts()
```

### 6 - clean ` gender` column 
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
# print('Before')
# print(df['gender'].unique())
# print(df['gender'].value_counts())
# print(df['gender'].isna().sum())

df['gender'] = df['gender'].str.lower().str.strip()
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])

# print('After')
# print(df['gender'].unique())
# df['gender'].value_counts()
```

### 7 - clean ` age` column 
> tip : note the type of age column , and NaN values
 - try to use `to_numeric` 
 - tip : you should choose which best value to put in place of these NaNs ( mean, median, ....etc)

```python
df['age'] = pd.to_numeric(df['age'] , errors='coerce')
df.loc[(df['age'] < 2) | (df['age'] > 100), 'age'] = np.nan
df['age'] = df['age'].fillna(df['age'].median()).astype('int')

df['age'].unique()
```

### 8 - clean ` country` column
> tip : note that country column has country names with different writing ways 
  - use fuzzywuzzy to deal with this problem
  - you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
from thefuzz import process
fix = {

    'united states': 'usa',
    'united staes': 'usa',
    'united stated': 'usa',
    'united sates': 'usa',
    'united ststes': 'usa',
    'unites states': 'usa',
    'unied states': 'usa',
    'united statss': 'usa',
    'the united states': 'usa',
    'u.s.a.': 'usa',
    'u s a': 'usa',
    'u.s.': 'usa',
    'u s': 'usa',
    'us of a': 'usa',
    'america': 'usa',
    'murica': 'usa',
    'murrika': 'usa',
    "'merica": 'usa',
    'trumpistan': 'usa', 
    'unhinged states': 'usa',
    'insanity lately': 'usa',
    'usa? hard to tell anymore..': 'usa',
    'ahem....amerca': 'usa',
    'united statea' : 'usa',
    
    # Specific US Locations to USA
    'new york': 'usa',
    'california': 'usa',
    'north carolina': 'usa',
    'pittsburgh': 'usa',
    'new jersey': 'usa',
    'alaska': 'usa',
    'cascadia': 'usa', 

    # UK Variations
    'uk': 'united kingdom',
    'u.k.': 'united kingdom',
    'england': 'united kingdom',
    'scotland': 'united kingdom',
    'endland': 'united kingdom',

    # Canada Variations
    'soviet canuckistan': 'canada',
    'canae': 'canada',
    'n. america': 'canada', 

    # Other
    'subscribe to dm4uz3 on youtube': np.nan,
    "i don't know anymore": np.nan,
    'earth': np.nan,
    'atlantis': np.nan,
    'narnia': np.nan,
    'fear and loathing': np.nan,
    'europe': np.nan,  
    '': np.nan
}
df['country'] = df['country'].str.lower().str.strip()
df['country'] = df['country'].replace(fix)

countries = [
    'usa', 'canada', 'united kingdom', 'france', 'germany', 'mexico', 
    'uae', 'finland', 'netherlands', 'australia', 'greece', 'south korea', 
    'japan', 'south africa', 'costa rica', 'denmark', 'switzerland', 
    'indonesia', 'singapore', 'china', 'taiwan', 'hong kong', 
    'spain', 'sweden', 'ireland',
]

for country_bad in df['country'].unique():
    if pd.notna(country_bad) and isinstance(country_bad, str) and country_bad not in countries: 
        match, score = process.extractOne(country_bad, countries)
        if score > 80:
            df.loc[df['country'] == country_bad, 'country'] = match
df['country'] = df['country'].fillna(df['country'].mode()[0]).astype('category')

df['country'].unique()
```
